```python

#Function for capping
data['H'] = np.where(
    data['H']>max,max,
    np.where(
        data['H']<min,min,
        data['H']
    )
)
```

Albedo column is skew data so we can use IQR method of otlier removal

```python
#Finding the IQR
percentile_25 = data['albedo'].quantile(0.25)
percentile_75 = data['albedo'].quantile(0.75)
IQR = percentile_75 -percentile_25
IQR
```

```python
upper = percentile_75 + 1.5*IQR
lower = percentile_25 - 1.5*IQR
print("Upper limit : ",upper)
print("Lower limit : ",lower)
```

```python
# data which has outliers in albedo
data[(data['albedo']>upper) | (data['albedo']<lower )]
```

```python
#Function for capping
data['albedo'] = np.where(
    data['albedo']>upper,
    upper,
    np.where(
        data['albedo']<lower,
        lower,
        data['albedo']
    )
)
```
